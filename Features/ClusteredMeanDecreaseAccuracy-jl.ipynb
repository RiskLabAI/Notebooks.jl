## Fit

```julia
function clusteredFeatureImportanceMDA(classifier, X::DataFrame, y::DataFrame, clusters, nSplits::Int64)
    cvGenerator = ModelSelection.KFold(n_splits=nSplits)    
    score0, score1 = zeros(nSplits), DataFrame([name => zeros(nSplits) for name in keys(clusters)])
    for (i, (train, test)) ∈ enumerate(cvGenerator.split(X |> Matrix))
        println("fold $(i) start ...")

        train .+= 1 # Python indexing starts at 0
        test .+= 1 # Python indexing starts at 0

        X0, y0 = X[train, :], y[train, :] 
        X1, y1 = X[test, :], y[test, :]

        fit = classifier.fit(X0 |> Matrix, y0 |> Matrix |> vec)

        predictionProbability = fit.predict_proba(X1 |> Matrix)
        score0[i] = -Metrics.log_loss(y1 |> Matrix, predictionProbability, labels=classifier.classes_)        
        rng = MersenneTwister(1234);
        for j ∈ keys(clusters)
            X1_ = deepcopy(X1) 
            for k ∈ clusters[j]
                X1_[:, names(X)[k]] = shuffle(rng, X1_[:, names(X)[k]])
            end
            predictionProbability = fit.predict_proba(X1_ |> Matrix)
            log_loss = Metrics.log_loss(y1 |> Matrix, predictionProbability, labels=classifier.classes_)
            score1[i, j] = -log_loss
        end
    end

    importances = .-score1 .+ score0
    importances = importances ./ .-score1

    means = mean.(eachcol(importances))
    stds = std.(eachcol(importances)) .* size(importances)[1] ^ -0.5
    indices = keys(clusters) |> collect

    results = DataFrame("ClusterIndex"=>indices, "Mean"=>means, "StandardDeviation"=>stds)
    return results
end 

classifier = Tree.DecisionTreeClassifier(
    criterion="entropy",
    max_features=1,
    class_weight="balanced",
    min_weight_fraction_leaf=0,)

classifier = Ensemble.BaggingClassifier(
    base_estimator=classifier,
    n_estimators=500,
    max_features=1.0,
    max_samples=1.0,
    oob_score=false)

results = clusteredFeatureImportanceMDA(classifier, X, y, clusters, 10,)
```

# Plot Results

```julia
templates.default = "plotly_dark";
PlotlyJS.templates

toSavePlot = plot(
    bar(
        results,
        x=:Mean,
        y=:ClusterIndex,
        error_x=attr(type="data", array=:StandardDeviation, visible=true),
        orientation="h",),
    PlotlyJS.Layout(
        title="MDA Results",
        width=800, height=600,
        xaxis_title="Feature Importance",
        yaxis_title="Cluster Index",
    )
)
```
